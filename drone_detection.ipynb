```python
# THIS CODE BUILDS OFF OF THE YOLOv8 MODEL
# Open-source github: https://github.com/doguilmak/Drone-Detection-YOLOv8x/blob/main/README.md
# Uses pre-trained weights from YOLO epoch 50 size: https://huggingface.co/doguilmak/Drone-Detection-YOLOv8x/tree/main/weight
# reference video https://www.youtube.com/watch?v=Bzv58L6xYGc
# previous approach was a classifier, not detecter model: https://www.youtube.com/watch?v=6YFvMeVi3Uc using https://teachablemachine.withgoogle.com/train/image

from ultralytics import YOLO
import cv2
import os

# Path to your pre-trained YOLOv8 model
model_path = "best.pt"  # make sure best.pt is in the same folder
if not os.path.exists(model_path):
    raise FileNotFoundError("best.pt not found in folder")

# Load the model
model = YOLO(model_path)

# Start webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Could not open webcam")

print("Starting live drone detection. Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)

    # Draw bounding boxes on frame
    annotated_frame = results[0].plot()

    # Show live feed
    cv2.imshow("Drone Detection", annotated_frame)

    # Quit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
